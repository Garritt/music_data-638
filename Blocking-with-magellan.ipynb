```python
## Explicitly state the type for each of the rows. 
MSD_TABLE["danceability"] = MSD_TABLE["danceability"].astype(str)
MSD_TABLE["danceability.1"] = MSD_TABLE["danceability.1"].astype(str)
MSD_TABLE["keySignature"] = MSD_TABLE["keySignature"].astype(str)
MSD_TABLE["tempo"] = MSD_TABLE["tempo"].astype(str)
MSD_TABLE["timeSignature"] = MSD_TABLE["timeSignature"].astype(str)
MSD_TABLE["title"] = MSD_TABLE["title"].astype(str)
MSD_TABLE["year"] = MSD_TABLE["year"].astype(str)
```

```python
# Rule-based blocking over input tables
# first get features that can be used
feature_table = em.get_features_for_blocking(MSD_TABLE, MBRAINZ_TABLE)
feature_table
```

```python
## Upon viewing the profiles for both tables above as well as the feature table, we will proceed
## with blocking on the "Title" attribute from MSD_TABLE and the 
## "Song" attribute from MBRAINZ_TABLE. Both of these attributes have
## a high uniqueness which will provide useful when blocking
```

```python
## We will have two different tokenizers to compare blocking results.
## The first is a tokenization by whitespace while the second is a 
## 3-gram tokenization. Both will be used with the Jacaard score to
## determine how similar the tuples are.
ws = sm.WhitespaceTokenizer(return_set=True)
gram3 = sm.QgramTokenizer(qval=3)
```

```python
## Using the whitespace tokenization, we will set our similarity 
## score threshold to be 0.5. This means that the tokens will have to
```

```python
## share at least 50% of the same words in order to be considered for
## the matching phase
```

```python
output_pairs_ws = ssj.jaccard_join(MSD_TABLE, MBRAINZ_TABLE, 'ID', 'ID', 'title', 'title', ws, 0.5, 
                                l_out_attrs=['title'], r_out_attrs=['title'])
```

```python
## The length of this list will be the solution space that will be 
## searched in the matching phase
len(output_pairs_ws)
```

```python
## We will rerun the blocking phase with a new threshold of 0.7 
## in attempts to reduce space even further
output_pairs_ws_7 = ssj.jaccard_join(MSD_TABLE, MBRAINZ_TABLE, 'ID', 'ID', 'title', 'title', ws, 0.7, 
                                l_out_attrs=['title'], r_out_attrs=['title'])
```

```python
len(output_pairs_ws_7)
```

```python
## This is the most reduced thus far
```

```python
## We will repeat the Jacaard score using the qgram technique
## this time. 
output_pairs_js_5 = ssj.jaccard_join(MSD_TABLE, MBRAINZ_TABLE, 'ID', 'ID', 'title', 'title', gram3, 0.5, 
                                l_out_attrs=['title'], r_out_attrs=['title'])
```

```python
len(output_pairs_js_5)
```

```python
output_pairs_js_7 = ssj.jaccard_join(MSD_TABLE, MBRAINZ_TABLE, 'ID', 'ID', 'title', 'title', gram3, 0.7, 
                                l_out_attrs=['title'], r_out_attrs=['title'])
```

```python
len(output_pairs_js_7)
```

```python
## We will print out all of the possible solution sets found by 
## blocking
print("Jacaard with whitespace tokenization and threshold of 0.5: ", len(output_pairs_ws))
print("Jacaard with whitespace tokenization and threshold of 0.7: ", len(output_pairs_ws_7))
print("Jacaard with 3-gram tokenization and threshold of 0.5: ", len(output_pairs_js_5))
print("Jacaard with 3-gram tokenization and threshold of 0.7: ", len(output_pairs_js_7))
```

```python
## Jacaard score using 3-gram and a threshold of 0.7 reduces the 
## possible solution space the most. We will use this table for 
## matching. 
```

```python
output_pairs_js_7
```
