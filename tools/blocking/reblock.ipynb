```python
import py_entitymatching as em
import pandas as pd
import os
```

```python
######### READING IN TABLES ############

path_a = os.path.join(os.getcwd(), "../../csv/msd_final.csv" )
path_b = os.path.join(os.getcwd(), "../../csv/musicbrainz_final.csv")

A = em.read_csv_metadata(path_a, key='ID')
B = em.read_csv_metadata(path_b, key='ID')

#########################################
```

```python
########### FEATURE RULES ##############

## Explicitly state the type for each of the rows. 
A["danceability"] = A["danceability"].astype(str)
A["duration"] = A["duration"].astype(str)
A["keysignature"] = A["keysignature"].astype(str)
A["tempo"] = A["tempo"].astype(str)
A["timesignature"] = A["timesignature"].astype(str)
A["title"] = A["title"].astype(str)
A["year"] = A["year"].astype(str)

feature_table = em.get_features_for_blocking(A, B)
feature_table

########################################
```

```python
######### TWO RULE BLOCKERS ################

rbt = em.RuleBasedBlocker()
rba = em.RuleBasedBlocker()

#rule_title = rbt.add_rule(['title_title_jac_qgm_3_qgm_3(ltuple, rtuple) < 0.5'], feature_table)
rule_title = rbt.add_rule(['title_title_jac_qgm_3_qgm_3(ltuple, rtuple) < 0.7'], feature_table)

#rule_artist = rba.add_rule(['artist_artist_jac_qgm_3_qgm_3(ltuple, rtuple) < 0.5'], feature_table)

candidates_t = rbt.block_tables(A, B, l_output_attrs=['title','artist','year'], r_output_attrs=['title','artist','year'])
#candidates_a = rba.block_tables(A, B, l_output_attrs=['title','artist','year'], r_output_attrs=['title','artist','year'])

candidates_t

#############################################
```

```python

### reduce candidate set size to candidates_final ###
cf = open('../../csv/candidates_final.csv', 'w')
cf.write("id,ltable_ID,rtable_ID,ltable_title,ltable_artist,ltable_year,rtable_title,rtable_artist,rtable_year\n")
ID = 0
for i, row in candidates_t.iterrows():
    lt_a = row['ltable_artist']
    rt_a = row['rtable_artist']
    
    if lt_a[0] == rt_a[0]:
        cf.write(str(ID)+","+row['ltable_ID']+","+row['rtable_ID']+","+row['ltable_title']+","+row['ltable_artist']+","+row['ltable_year']+","+row['rtable_title']+","+row['rtable_artist']+","+row['rtable_year']+"\n")
        ID+=1
        
cf        
```
